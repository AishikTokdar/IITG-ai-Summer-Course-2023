### 5) Print the unique values in each column

```python
for column in a.columns:
    unique_values = a[column].unique()
    print(f"Column '{column}' unique values:")
    print(unique_values)
    print()
```

### 6) Find the columns that has null values in them

Hint: You can use isnull() function

```python
columns_with_null_values = a.columns[a.isnull().any()]
print("Columns with null values:")
for column in columns_with_null_values:
    print(column)
```

### 7) Show various statistics of all the numerical columns

Hint: Use describe()

```python
numerical_columns = a.select_dtypes(include=[float, int])
statistics = numerical_columns.describe()
print(statistics)
```

### 8) Drop the column that has maximum null values and update the Dataframe

```python
column_with_max_nulls = a.isnull().sum().idxmax()
a = a.drop(column_with_max_nulls, axis=1)
print(a)
```

### 9) Display the data type of the remaining columns(use updated dataframe)

### 10) Display the sex ratio (female /male)
use value_counts()

```python
sex_counts = a['Sex'].value_counts()
female_count = sex_counts['female']
male_count = sex_counts['male']
sex_ratio = female_count / male_count
print("Sex Ratio (Female / Male):", sex_ratio)
```

### 11) Find the number of people that Survived

```python
survived_counts = a['Survived'].value_counts()
num_survived = survived_counts[1]
print("Number of people who survived:", num_survived)
```

### 12) Find the person with maximum and minimum age
use .max() and .min() <br>
You need to print the row of the dataframe where the person with max and min age is present

```python
max_age = a['Age'].max()
min_age = a['Age'].min()
```

```python
person_with_max_age = a[a['Age'] == max_age]
person_with_min_age = a[a['Age'] == min_age]

print("Person with maximum age:\n", person_with_max_age)
print("\nPerson with minimum age:\n", person_with_min_age)
```

### 13) Find the number of Class 2 tickets sold

```python
class_counts = a['Pclass'].value_counts()
num_class_2_tickets_sold = class_counts[2]
print("Number of Class 2 tickets sold:", num_class_2_tickets_sold)
```

### 14) Find the number of females who survived the crash

```python
num_females_survived = len(a[(a['Survived'] == 1) & (a['Sex'] == 'female')])
print("Number of females who survived:", num_females_survived)
```

### 15) Find the number of people who survived the crash and who paid fare less than 200

```python
num_people_survived_low_fare = len(a[(a['Survived'] == 1) & (a['Fare'] < 200)])
print("Number of people who survived and paid fare less than 200:", num_people_survived_low_fare)
```
